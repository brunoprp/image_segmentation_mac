```python
def trackChaned(x):
    pass
 
cv2.namedWindow('Color Track Bar')

cv2.createTrackbar('lowH', 'Color Track Bar', 0,255, trackChaned)
cv2.createTrackbar('highH', 'Color Track Bar', 0,255, trackChaned)

cv2.createTrackbar('lowS', 'Color Track Bar', 0,255, trackChaned)
cv2.createTrackbar('highS', 'Color Track Bar', 0,255, trackChaned)

cv2.createTrackbar('lowV', 'Color Track Bar', 0,255, trackChaned)
cv2.createTrackbar('highV', 'Color Track Bar', 0,255, trackChaned)
captura = cv2.VideoCapture('http://10.0.0.111:4747/video')
#captura = cv2.VideoCapture(0)


while(True):
    ret, frame = captura.read()
    
    
    
    ilowH = cv2.getTrackbarPos('lowH', 'Color Track Bar')
    ihighH = cv2.getTrackbarPos('highH', 'Color Track Bar')
    
    ilowS = cv2.getTrackbarPos('lowS', 'Color Track Bar')
    ihighS = cv2.getTrackbarPos('highS', 'Color Track Bar')
    
    ilowV = cv2.getTrackbarPos('lowV', 'Color Track Bar')
    ihighV = cv2.getTrackbarPos('highV', 'Color Track Bar')
    
    
    lower_hsv = np.array([ilowH, ilowS, ilowV])
    higher_hsv = np.array([ihighH, ihighS, ihighV])
    
    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    skinMask = cv2.inRange(frame_hsv, lower_hsv, higher_hsv)
    
    # aplicando uma srie de eroses e dilataes na mscara usando elliptical kernel
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (11, 11))
    skinMask = cv2.erode(skinMask, kernel, iterations = 2)
    skinMask = cv2.dilate(skinMask, kernel, iterations = 2)

    # Desfocando a mscara para ajudar a remover o rudo 
    skinMask = cv2.GaussianBlur(skinMask, (3, 3), 0)
    
    skin = cv2.bitwise_and(frame, frame, mask = skinMask)
    
    
    
    cv2.imshow("frame", frame)
    cv2.imshow("mask Skin", skin)
    cv2.imshow("mask", skinMask)
    
    # Esc para sair 
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
captura.release()
cv2.destroyAllWindows()
```
