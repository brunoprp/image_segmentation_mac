# Morphological Snakes
* Existem dois métodos *Morphological Snakes* disponíveis no Skimage: *Morphological Geodesic Active Contours* ( implementado na função *morphological_geodesic_active_contour*) e *Morphological Active Contours without Edges* ( implementado na função *morphological_chan_vese*).

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float
from skimage.segmentation import (morphological_chan_vese,
                                  morphological_geodesic_active_contour,
                                  inverse_gaussian_gradient,
                                  checkerboard_level_set)
```

```python
def store_evolution_in(lst):
    """Returns a callback function to store the evolution of the level sets in
    the given list.
    """

    def _store(x):
        lst.append(np.copy(x))

    return _store
```

## 1) Morphological Geodesic Active Contours - MorphGAC (Contornos ativos geodésicos morfológicos)

* A MorphGAC é adequado para imagens com contornos visíveis, mesmo quando esses contornos podem ser ruidosos, desordenados ou parcialmente confusos. Requer, no entanto, que a imagem seja pré-processada para destacar os contornos. Isso pode ser feito por exemplo usando a função *inverse_gaussian_gradient*. A qualidade da segmentação MorphGAC depende muito dessa etapa de pré-processamento.

```python

# Morphological GAC
image_GAC = img_as_float(data.coins())
gimage = inverse_gaussian_gradient(image_GAC)
```

```python
plt.figure(figsize=(13, 13))
plt.subplot(2,2,1),plt.imshow(image_GAC,cmap = 'gray')
plt.title('Image Gray'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(gimage,cmap = 'gray')
plt.title('Inverse gaussian gradient'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Initial level set
init_ls_GAC = np.zeros(image_GAC.shape, dtype=np.int8)
init_ls_GAC[10:-10, 10:-10] = 1

# List with intermediate results for plotting the evolution
evolution_GAC = []
callback_GAC = store_evolution_in(evolution_GAC)
ls_GAC = morphological_geodesic_active_contour(gimage, iterations=230,
                                           init_level_set=init_ls_GAC,
                                           smoothing=1, balloon=-1,
                                           threshold=0.69,
                                           iter_callback=callback_GAC)
```

## 2) Morphological Active Contours without Edges - MorphACWE (Contornos Morfológicos Ativos sem Bordas)
* Ao contrário, a MorphACWE funciona bem quando os valores de pixel das regiões internas e externas do objeto a segmentar têm médias diferentes. Ao contrário do MorphGAC , o MorphACWE não exige que os contornos do objeto sejam bem definidos e funciona sobre a imagem original sem nenhum processamento anterior. Isso torna o MorphACWE mais fácil de usar e ajustavel do que o MorphGAC .

```python
# Morphological ACWE
image_ACWE = img_as_float(data.camera())

# Initial level set
init_ls_ACWE = checkerboard_level_set(image_ACWE.shape, 6)
# List with intermediate results for plotting the evolution
evolution_ACWE = []
callback_ACWE = store_evolution_in(evolution_ACWE)
ls_ACWE = morphological_chan_vese(image_ACWE, iterations=35, init_level_set=init_ls_ACWE,
                             smoothing=3, iter_callback=callback_ACWE)
```

```python



fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.flatten()

# Morphological ACWE
ax[0].imshow(image_ACWE, cmap="gray")
ax[0].set_axis_off()
ax[0].contour(ls_ACWE, [0.5], colors='r')
ax[0].set_title("Morphological ACWE segmentation", fontsize=12)

ax[1].imshow(ls_ACWE, cmap="gray")
ax[1].set_axis_off()
contour = ax[1].contour(evolution_ACWE[2], [0.5], colors='g')
contour.collections[0].set_label("Iteration 2")
contour = ax[1].contour(evolution_ACWE[7], [0.5], colors='y')
contour.collections[0].set_label("Iteration 7")
contour = ax[1].contour(evolution_ACWE[-1], [0.5], colors='r')
contour.collections[0].set_label("Iteration 35")
ax[1].legend(loc="upper right")
title = "Morphological ACWE evolution"
ax[1].set_title(title, fontsize=12)


# Morphological GAC

ax[2].imshow(image_GAC, cmap="gray")
ax[2].set_axis_off()
ax[2].contour(ls_GAC, [0.5], colors='r')
ax[2].set_title("Morphological GAC segmentation", fontsize=12)

ax[3].imshow(ls_GAC, cmap="gray")
ax[3].set_axis_off()
contour = ax[3].contour(evolution_GAC[0], [0.5], colors='g')
contour.collections[0].set_label("Iteration 0")
contour = ax[3].contour(evolution_GAC[100], [0.5], colors='y')
contour.collections[0].set_label("Iteration 100")
contour = ax[3].contour(evolution_GAC[-1], [0.5], colors='r')
contour.collections[0].set_label("Iteration 230")
ax[3].legend(loc="upper right")
title = "Morphological GAC evolution"
ax[3].set_title(title, fontsize=12)

fig.tight_layout()
plt.show()
```
